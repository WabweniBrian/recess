```python
#Decision Tree
#import libraries
import numpy as np #linear algebra
import pandas as pd
#import graphviz # visulaozation of the leaf  nodes
import seaborn as sns  #statistical data visualization
import matplotlib.pyplot as plt
%matplotlib inline

#ML Libraries
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

#control display |Optional

```

```python
df = pd.read_csv("heart_disease.csv")
df.head(20)
```

```python
#show number of people with heart disease
#column target 0-for not having the heart disease and 1 for having the heart disease
df.target.value_counts()
```

```python
#show percentages of male and female that  have heart diseases
count_female = len(df[df.sex==0])
count_male = len(df[df.sex==1])
print( 'percentage of female patient : {:.2f}%'.format(count_female/(len(df.sex))*100))
print( 'percentage of male patient : {:.2f}%'.format(count_male/(len(df.sex))*100))
```

```python
#show the mean
df.groupby('target').mean()
```

```python
#show for null or nan values
df.isnull().sum()
```

```python
#ML train and test data
#create logistic regression
y = df.target.values
X_data = df.drop(['target'],axis=1)
```

```python
#split the data for training 67% and testing 33%
X_train,X_test,y_train,y_test = train_test_split(X_data,y,test_size=0.33,random_state=42)
```

```python
#check shape of data remaining after splitting for training and testing
X_train.shape, X_test.shape
```

```python
#check the datatype
X_train.dtypes
```

```python
X_train.head()
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini',  max_depth=3, random_state=0)
clf_gini.fit(X_train, y_train)
```

```python
#predict the class label for the test data using the decision tree classifier
y_pred_gini = clf_gini.predict(X_test)

from sklearn.metrics import accuracy_score
print('Model accuracy score with criterion gini index:{0:0.4}'.format(accuracy_score(y_test,y_pred_gini)))
```

```python
#predict the class label for training data using decision tree classifier
y_pred_train_gini = clf_gini.predict(X_train)
y_pred_train_gini
```

```python
#print the training set of accuracy score
print('Training-set accuracy score:{0:0.4f}'.format(clf_gini.score(X_train, y_train)))
print('Testing-set accuracy score:{0:0.4f}'.format(clf_gini.score(X_test, y_test)))
```

```python
#1st way to visual Decision Tree Leaf Node
plt.figure(figsize=(12,8)) #width and height
from sklearn import tree
tree.plot_tree(clf_gini.fit(X_train,y_train))
```
